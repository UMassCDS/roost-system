# great lakes

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
## MODEL PREDICTIONS, all years
n_seg = 20

side = 600./n_seg
plt.figure(figsize=(3,3))
data = np.zeros((n_seg, n_seg))
for file in sorted(os.listdir("roostui/data/all_stations_v2_screened")):
    if not file.startswith("roost_labels"): continue    
    annotations = [annotation.strip().split(",") for annotation in open(
        "roostui/data/all_stations_v2_screened/" + file, "r"
    ).readlines()[1:]]
    for annotation in annotations:
        if annotation[21] == "swallow-roost":
            data[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
data /= np.sum(data)
plt.title("model predictions, all station years")
plt.imshow(data)
plt.show()
```

```python
## MODEL PREDICTIONS, given a station year, 5 years, 25 locations vs. number of roosts
n_seg = 20

side = 600./n_seg
station = None
ii = -1
jj = -1
plt.figure(figsize=(15,36))
for file in sorted(os.listdir("roostui/data/all_stations_v2_screened")):
    if not file.startswith("roost_labels") or int(file.split("_")[3][:4]) not in [2000, 2005, 2010, 2015, 2020]: continue
    current_station = file.split('_')[2]
    if current_station != station:
        ii += 1
        jj = 0
        station = current_station
    else:
        jj += 1
    
    data = np.zeros((n_seg, n_seg))
        
    annotations = [annotation.strip().split(",") for annotation in open(
        "roostui/data/all_stations_v2_screened/" + file, "r"
    ).readlines()[1:]]
    for annotation in annotations:
        if annotation[21] == "swallow-roost":
            data[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
    data /= np.sum(data)

    ax = plt.subplot(12, 5, 5*ii+jj+1)
    ax.imshow(data)
#     for i in range(n_seg):
#         for j in range(n_seg):
#             text = ax.text(j, i, f"{data[i, j]:.2f}", ha="center", va="center", color="w")
    ax.set_title(f"{current_station}_{file.split('_')[3][:4]}")
    
# fig.tight_layout()
plt.show()
```

```python
## Given a station year, 25 locations vs. precision, precision
n_seg = 20

side = 600./n_seg
station = None
ii = -1
jj = -1
plt.figure(figsize=(15,36))
for file in sorted(os.listdir("roostui/data/all_stations_v2_screened")):
    if not file.startswith("roost_labels") or int(file.split("_")[3][:4]) not in [2000, 2005, 2010, 2015, 2020]: continue
    current_station = file.split('_')[2]
    if current_station != station:
        ii += 1
        jj = 0
        station = current_station
    else:
        jj += 1
    
    data = np.zeros((n_seg, n_seg))
    correct = np.zeros((n_seg, n_seg))
        
    annotations = [annotation.strip().split(",") for annotation in open(
        "roostui/data/all_stations_v2_screened/" + file, "r"
    ).readlines()[1:]]
    for annotation in annotations:
        if annotation[21] == "swallow-roost":
            data[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
        if annotation[20] in ["swallow-roost", "weather-roost", "unknown-noise-roost", "AP-roost"] and \
                annotation[21] == "swallow-roost":
            correct[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
    data = correct / data

    ax = plt.subplot(12, 5, 5*ii+jj+1)
    ax.imshow(data)
#     for i in range(n_seg):
#         for j in range(n_seg):
#             text = ax.text(j, i, f"{data[i, j]:.2f}", ha="center", va="center", color="w")
    ax.set_title(f"{current_station}_{file.split('_')[3][:4]}")
    
# fig.tight_layout()
plt.show()
```

```python
## Precision, all years
n_seg = 20

side = 600./n_seg
plt.figure(figsize=(3,3))
correct = np.zeros((n_seg, n_seg))
data = np.zeros((n_seg, n_seg))
for file in sorted(os.listdir("roostui/data/all_stations_v2_screened")):
    if not file.startswith("roost_labels"): continue    
    annotations = [annotation.strip().split(",") for annotation in open(
        "roostui/data/all_stations_v2_screened/" + file, "r"
    ).readlines()[1:]]
    for annotation in annotations:
        if annotation[21] == "swallow-roost":
            data[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
        if annotation[20] in ["swallow-roost", "weather-roost", "unknown-noise-roost", "AP-roost"] and \
                annotation[21] == "swallow-roost":
            correct[int(float(annotation[4])/side), int(float(annotation[5])/side)] += 1
data = correct / data
plt.title("precision, all station years")
im = plt.imshow(data)

plt.colorbar()
plt.show()
print()
```

# v0.1.0 test

```python
import os, json
exp = "roost-system/development/custom_eval/3.2eval_resnet101-FPN_detptr_anc10_regsl1_imsz1200_lr0.001_it150k_ckpt39999_strt1/"
```

```python
with open(os.path.join(exp, "roosts_v0.1.0_test_coco_format.json"), "r") as f:
    gt_ = json.load(f)["annotations"]

n_seg = 20

side = 600./n_seg
plt.figure(figsize=(3,3))
data = np.zeros((n_seg, n_seg))
for gt in gt_:
    bbox = gt["bbox"]
    try:
        data[int((bbox[0]+bbox[2]/2)/side), int((bbox[1]+bbox[3]/2)/side)] += 1
    except:
        print(int((bbox[0]+bbox[2]/2)/side), int((bbox[1]+bbox[3]/2)/side))
data /= np.sum(data)
plt.title("ground truth")
plt.imshow(data)
plt.show()
```

```python
with open(os.path.join(exp, "coco_instances_results.json"), "r") as f:
    gt_ = json.load(f)
    
n_seg = 20

side = 600./n_seg
plt.figure(figsize=(3,3))
data = np.zeros((n_seg, n_seg))
for gt in gt_:
    bbox = gt["bbox"]
    try:
        data[int((bbox[0]+bbox[2]/2)/side), int((bbox[1]+bbox[3]/2)/side)] += 1
    except:
        print(int((bbox[0]+bbox[2]/2)/side), int((bbox[1]+bbox[3]/2)/side))
data /= np.sum(data)
plt.title("model prediction")
plt.imshow(data)
plt.show()
```
